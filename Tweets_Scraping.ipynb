### 2) Scraping tweets about Pfizer Vaccine

```python
def tweet_scraping(keyword,lang,timeOut,output_fileName):
  """
    This method will achieve the scraping of tweets where the research is based on the keywords given by the user 
    
    :input:

      keyword: topics to search for, hashtags & trends ..
      lang: the written language of tweets
      output_fileName : the csv file name to store scraped data
      timeOut : number of seconds chosen to make the research

    :output: 

      csv file
  """
  #configuration

  c = twint.Config()
  c.Search = keyword
  c.Store_csv = True
  c.Output = output_fileName
  c.Lang = lang

  #Run Search
  twint.run.Search(c)
  """
  from interruptingcow import timeout #must be reimported
  try:
    with timeout(timeOut, exception=RuntimeError):
      while True:
        twint.run.Search(c)
  except RuntimeError:
    pass
  """  
```

```python
keywords= ["pfizer vaccine","PfizerVaccine"]
timeout= 10 # [seconds]
language="en"
```

```python
#Search 1 : with a topic
tweet_scraping(keywords[0],language,timeout,"Pfizer_Tweets1.csv")
```

```python
#Search 2 : with a hashtag
tweet_scraping(keywords[1],language,timeout,"Pfizer_Tweets2.csv")
```

```python
df1=pd.read_csv('Pfizer_Tweets1.csv')
df2=pd.read_csv('Pfizer_Tweets2.csv')
print(df1.shape,'\n',df2.shape)
```

```python
#Combining the two csv files 
csv_file=open("./Pfizer_Tweets.csv", mode='w',newline='',encoding="utf-8")
fieldnames = ['id', 'tweet']
writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
writer.writeheader()

for i in range(len(df1['id'])):
    writer.writerow({'id': df1['id'][i] , 'tweet':df1['tweet'][i]})
for i in range(len(df2['id'])):
    writer.writerow({'id': df2['id'][i] , 'tweet':df2['tweet'][i]})

df = pd.read_csv("./Pfizer_Tweets.csv")
```
